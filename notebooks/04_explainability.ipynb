# Stage 4: Model Explainability

**Objective:** Interpret the trained LightGBM model to:
1. **Globally** — understand which features drive risk across the entire portfolio.
2. **Locally** — explain individual loan decisions for specific applicants.
3. **Build trust** — satisfy regulatory requirements (ECOA, FCRA, Basel IRB).

**Model:** LightGBM Classifier (AUC 0.7680, 182 features)
**Data:** 20% holdout test set (~60K applicants)

---

## Analysis Plan

### 1. Global Interpretation

| Method | Purpose |
|---|---|
| Native Feature Importance | Fast signal of gain-based importance from trained model |
| SHAP Summary (Beeswarm) | Signed, unbiased feature impact with direction and spread |
| SHAP Bar Plot | Precise mean |SHAP| ranking for regulatory reporting |
| Partial Dependence Plots | Reveal shape of relationship; translate to policy thresholds |
| Permutation Importance | AUC drop when feature is shuffled; quantifies data source value |

### 2. Local Interpretation

| Method | Purpose |
|---|---|
| SHAP Waterfall (High-Risk) | Decompose prediction for the riskiest applicant |
| SHAP Waterfall (Low-Risk) | Contrast with the safest applicant |
| SHAP Force Plot | Executive-friendly alternative visualisation |

### 3. Model Monitoring Framework

Seven monitoring dimensions: Data Drift, Feature Drift (PSI), Prediction Drift,
Population Stability, Concept Drift, Model Performance, Calibration.

---

## Expected Findings

1. **Top risk drivers:** EXT_SOURCE_2, DTI, PREV_DEFAULT_RATE, LTI, AGE_YEARS
2. **Model is economically rational** — matches human underwriting intuition.
3. **Actionable thresholds** from PDPs (DTI > 0.40 triples risk).
4. **Individual explanations** satisfy regulatory adverse-action requirements.

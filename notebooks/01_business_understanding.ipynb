# Stage 1: Business Understanding

**Project:** Credit Risk Analytics — Horizon Lending Inc.
**Target:** TARGET (1 = default, 0 = repaid, ~8% default rate)

---

## Business Problem

Horizon Lending Inc. is a mid-sized digital consumer-lending platform offering unsecured personal loans ($1,000–$35,000, 12–60 month terms) to near-prime borrowers (credit scores 580–680). This segment is underserved by traditional banks but carries elevated default risk.

**Core problem:** Each percentage-point improvement in default prediction accuracy translates to millions in avoided losses. Horizon needs a data-driven model to:
- Identify high-risk applicants before funding.
- Reduce net charge-offs by 15–20%.
- Price loans appropriately for each risk segment.
- Prioritise collections on accounts most likely to default.

**Business impact:** A 5-point AUC improvement (0.75 → 0.80) is estimated to save $2–3M annually.

| Table | Rows | Description |

| `installments_payments` | ~13.6M | Payment history on past loans |
| `POS_CASH_balance` | ~10M | POS/cash loan balances |
| `credit_card_balance` | ~3.8M | Credit card balances |

---

## Business Objectives

1. **Primary:** Build a default-probability model with AUC ≥ 0.80 on holdout test set.
2. **Secondary:** Translate model outputs into dollar-denominated profit simulations.
3. **Tertiary:** Produce an interactive dashboard for stakeholders.
4. **Compliance:** Perform fairness analysis to ensure no disparate impact.

---

## Stakeholder Map

| Stakeholder | Concern |
|---|---|
| Chief Risk Officer | Is the model sound, fair, and compliant? |
| Underwriting Team | Can we trust this score to automate decisions? |
| Collections Manager | Which accounts should we call first? |
| Finance / Treasury | What is our expected loss next quarter? |
| Regulatory Compliance | Does the model discriminate? |
| Product Manager | Can we offer better rates to good borrowers? |

## Success Criteria

| Metric | Target |
|---|---|
| ROC-AUC | ≥ 0.80 |
| Gini Coefficient | ≥ 0.60 |
| KS Statistic | ≥ 0.45 |
| Precision@5% | ≥ 0.40 |
| Calibration (Brier) | < 0.08 |
